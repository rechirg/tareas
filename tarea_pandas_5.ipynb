```python
import pandas as pd


df=pd.read_csv(r'C:\bootcamo\schools_complete.csv')

df2=pd.read_csv(r'C:\bootcamo\students_complete.csv')

table1=pd.merge(df2,df, on='school_name',how='left')

bins = [0,64,100]
bins2=[0,56,100]

group_names = ["Fail", "Pass"]

table1["Grade reading"]= pd.cut(table1["reading_score"], bins, labels=group_names)

table1["Grade math"]= pd.cut(table1["math_score"], bins2, labels=group_names)

table_pass_r=table1[table1["reading_score"]>64]
table_pass_r

table_pass_m=table1[table1["math_score"]>56]
table_pass_m

tablepassr=pd.pivot_table(table_pass_r,index=["school_name","type"],values=['Student ID'],aggfunc='count')
tablepassr

tablepassm=pd.pivot_table(table_pass_m,index=["school_name","type"],values=['Student ID'],aggfunc='count')
tablepassm


pass_reading=pd.pivot_table(table1,index=["school_name","type","Grade reading"],values=['Student ID'],aggfunc='count')


conteo_student=pd.pivot_table(table1,index=["school_name","type"],values=['Student ID'],aggfunc='count')

budget=pd.pivot_table(table1,index=["school_name","type"],values=['budget'],aggfunc='sum')

table2=pd.merge(conteo_student,budget, on=["school_name","type"],how='left')


table2['Per Student Budget']=table2['budget']/table2['Student ID']

avg_math=pd.pivot_table(table1,index=["school_name","type"],values=['math_score'],aggfunc='mean')

avg_math

table3=pd.merge(table2,avg_math, on=["school_name","type"],how='left')
table3

avg_reading=pd.pivot_table(table1,index=["school_name","type"],values=['reading_score'],aggfunc='mean')
avg_reading

table4=pd.merge(table3,avg_reading, on=["school_name","type"],how='left')
table4

table5=pd.merge(table4,tablepassr,on=["school_name","type"],how='left')
table5

table6=pd.merge(table5,tablepassm,on=["school_name","type"],how='left')
table6

table6['% Pass Reading']=table6['Student ID_y']/table6['Student ID_x']
table6

table6['% Pass Math']=table6['Student ID']/table6['Student ID_x']
table6

table7=table6.nlargest(5, columns=["% Pass Reading","% Pass Math","Student ID_x"])
table7
```
