```python
import pandas as pd


df=pd.read_csv(r'C:\bootcamo\purchase_data.csv')

pivot=pd.pivot_table(df,index=["Item ID"],values=['Price'],aggfunc='sum')
 
pivot



result=pivot.sort_values(('Price'),ascending=False)

result

top5=result.nlargest(5, columns='Price')

top5
```

```python
pivot2=pd.pivot_table(df,index=["Item ID"],values=['Price'],aggfunc='max')



pivot3=pd.pivot_table(df,index=["Item ID"],values=['Price'],aggfunc='min')

pivot4=pd.pivot_table(df,index=["Item ID"],values=['Price'],aggfunc='mean')


join=pd.merge(top5,pivot2, on='Item ID',how='left')


join3=pd.merge(join,pivot3, on='Item ID',how='left')

join4=pd.merge(join3,pivot4, on='Item ID',how='left')

join4
```
