```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pylab import *
from scipy import *


df1=pd.read_csv(r'C:\bootcamo\ride_data.csv')
df2=pd.read_csv(r'C:\bootcamo\city_data.csv')

t1 = pd.merge(df1,df2, how='left')

t2=t1.set_index('city')


```

```python
t3=pd.pivot_table(t2,index=["city"],values=['fare'],aggfunc='mean')


t4=pd.pivot_table(t2,index=["city"],values=['ride_id'],aggfunc='count')


t5=pd.pivot_table(t2,index=["city"],values=['driver_count'],aggfunc='sum')


u1= pd.merge(t3,t4, on='city',how='left')


u2=pd.merge(u1,t5, on='city',how='left')


```

```python
plt.scatter(u2.fare, u2.ride_id
, s=u2.driver_count,c=u2.fare)
plt.show()
```
